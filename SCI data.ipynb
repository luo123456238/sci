```python
import pandas as pd
```

```python
#打开本月数据
df_thismonth = pd.read_excel('./SCI202409.xls')
df_thismonth = df_thismonth[['Addresses','Reprint Addresses','Article Title','WoS Categories','Source Title','Volume','Issue','Start Page','End Page','DOI','UT (Unique WOS ID)']].rename(columns={'Addresses':'C1','Reprint Addresses':'RP','Article Title':'TI','WoS Categories':'WC','Source Title':'SO','Volume':'VL','Issue':'IS','Start Page':'BP','End Page':'EP','UT (Unique WOS ID)':'UT'})
df_thismonth.UT = df_thismonth.UT.astype(str)
#打开上月数据
df_lastmonth = pd.read_excel('./SCI202406.xlsx')
df_lastmonth = df_lastmonth[['C1','RP','TI','WC','SO','VL','IS','BP','EP','DI','UT']].rename(columns={'DI':'DOI'})
df_lastmonth.UT = df_lastmonth.UT.astype(str)
#合并去重
df = pd.concat([df_thismonth,df_lastmonth],axis=0).drop_duplicates(subset='UT',keep=False)
```

```python
df.to_excel('./SCI20240622-0921.xlsx')
```

```python
df = pd.read_excel('./11月-.xlsx')
df
```

```python
df = df[['C1','RP','TI','WC','SO','VL','IS','BP','EP','DI','UT']].rename(columns={'DI':'DOI'})
df
```

```python
#提取本校作者
df['pro']=df.C1.apply(lambda x:x.split('['))
df1 = df.explode('pro')
df2 = df1[df1.pro.str.contains('Fujian Med Univ') == 1]
df2 = df2.reset_index()
#去除姓名以外信息
df2['pro1'] = df2['pro'].apply(lambda x:str(x).split(']')[0])
#同一篇文章作者合并
df3 = df2.groupby(['UT']).agg({'pro1':[';'.join]})
df3 = df3.reset_index()
new_column = ['UT','name']
df3.columns = new_column
#同一篇文章作者去重
df3.name = df3.name.apply(lambda x:x.replace('; ',';'))
data_list = df3['name'].apply(lambda x:x.split(';'))
data_list = data_list.apply(lambda x:';'.join(set(x)))
df3['unique_name']=data_list
#合并数据
df4 = pd.merge(df,df3,on='UT')
df4.RP = df4.RP.str.split('(').str[0]
df4['BP'] = df4['BP'].astype(str)
df4['EP'] = df4['EP'].astype(str)
df4['页数']=df4['BP']+'-'+df4['EP']
p1 = df4[df4['页数']=='nan-nan']
p1.页数=p1.DOI
p2 = df4[df4['页数']!='nan-nan']
p3 = pd.concat([p1,p2])
p3= p3.rename(columns={'页数':'页数/DOI'})
p3
```
